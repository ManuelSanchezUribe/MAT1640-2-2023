```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(dpi=200)
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "monospace",
    "font.monospace": 'Computer Modern Typewriter'
})
```

```python
def concentracion_soluto(t, c0, s0):
    return c0 * (500 - 2 * t) + ((s0 - 500 * c0) / (500 ** 3)) * (500 - 2 * t) ** 3
```

```python
tiempo = np.linspace(0, 249, 500)
```

```python
fig, ax = plt.subplots()
ax.plot(tiempo, concentracion_soluto(tiempo, 5, 3), label='$x(t)$ con $c_{0}$ = 5, $s_{0}$ = 3')
ax.plot(tiempo, concentracion_soluto(tiempo, 10, 4), label='$x(t)$ con $c_{0}$ = 10, $s_{0}$ = 4')
ax.plot(tiempo, concentracion_soluto(tiempo, 1, 6), label='$x(t)$ con $c_{0}$ = 1, $s_{0}$ = 6')
ax.plot(tiempo, concentracion_soluto(tiempo, 20, 30), label='$x(t)$ con $c_{0}$ = 20, $s_{0}$ = 30')
plt.xlabel("Intervalo temporal")
plt.ylabel("Cantidad de soluto")
ax.set_title(r'\textbf{Tranques}')
ax.legend()
plt.savefig("Tranques.pdf")
plt.show()
```
